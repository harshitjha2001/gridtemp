```python
import numpy as np
```

```python
import pandas as pd
import pickle
```

```python
data = pd.read_csv('user_ratings.csv')
```

```python
data

```

```python
pt =data.pivot_table(index='Product',columns='User',values='Rating')
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommendx(prod):
    index = np.where(pt.index == prod)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)

    recommended_products = []
    count = 0 
    for i in similar_items:
        if pt.index[i[0]] != prod:  
            recommended_products.append(pt.index[i[0]])
            count += 1
            if count >= 3:  
                break
    
    return recommended_products

recommended_products = recommendx('HP Pavilion')
print("Recommended Products:")
for product in recommended_products:
    print(product)

```

```python
pickle.dump(pt,open('pt.pkl','wb'))
pickle.dump(similarity_scores,open('similarity_scores.pkl','wb'))
```

```python
pickle.dump(recommended_products,open('recommended_products.pkl','wb'))
```
