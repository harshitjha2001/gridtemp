```python
import numpy as np
```

```python
import pandas as pd
import pickle
```

```python
data1 = pd.read_csv('user_ratings.csv')
```

```python
data1

```

```python
avg_rating_df=data1.groupby('Product').mean()['Rating'].reset_index()
avg_rating_df.rename(columns={'Rating':'avg-ratings'},inplace=True)
avg_rating_df
```

```python
popular_df=avg_rating_df
popular_df


```

```python
popular_df=popular_df.sort_values('avg-ratings',ascending=False)
```

```python
popular_df=popular_df.head(9)
popular_df
```

```python
pickle.dump(popular_df,open('popular.pkl','wb'))
```
